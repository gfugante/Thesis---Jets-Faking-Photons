## Wmunu

```python
with open("R_pesati/Wmunu_0_70.txt", "r") as f:
    Wmunu_0_70 = yaml.load(f, Loader=yaml.SafeLoader)
with open("R_pesati/Wmunu_70_140.txt", "r") as f:
    Wmunu_70_140 = yaml.load(f, Loader=yaml.SafeLoader)
with open("R_pesati/Wmunu_140_280.txt", "r") as f:
    Wmunu_140_280 = yaml.load(f, Loader=yaml.SafeLoader)
with open("R_pesati/Wmunu_280_500.txt", "r") as f:
    Wmunu_280_500 = yaml.load(f, Loader=yaml.SafeLoader)
with open("R_pesati/Wmunu_500_1000.txt", "r") as f:
    Wmunu_500_1000 = yaml.load(f, Loader=yaml.SafeLoader)
with open("R_pesati/Wmunu_1000.txt", "r") as f:
    Wmunu_1000 = yaml.load(f, Loader=yaml.SafeLoader)
```

```python
print(Wmunu_0_70)
```

```python
plt.figure(num=None, figsize=(30, 12), dpi=80, facecolor='w', edgecolor='k')

x = []
y = []
yerr = []

# pt 0-70
for i in range(21):
    x.append(i)
    y.append(Wmunu_0_70[list(Wmunu_0_70.keys())[i]]["R_MC"])
    yerr.append(Wmunu_0_70[list(Wmunu_0_70.keys())[i]]["sigma_R_MC"])

plt.errorbar(x, y, yerr=yerr, fmt="ob", linestyle="-", linewidth=2)
    
# pt 70-140
for i in range(21):
    y[Wmunu_70_140[list(Wmunu_70_140.keys())[i]]["R_MC"]]
    yerr[Wmunu_70_140[list(Wmunu_70_140.keys())[i]]["sigma_R_MC"]]

plt.errorbar(x, y, yerr=yerr, fmt="og", linestyle="-", linewidth=2) 

# pt 140-280
for i in range(21):
    plt.errorbar(i, Wmunu_140_280[list(Wmunu_140_280.keys())[i]]["R_MC"], 
                 yerr=Wmunu_140_280[list(Wmunu_140_280.keys())[i]]["sigma_R_MC"], fmt="or", linestyle="-")

# pt 280-500
for i in range(21):
    plt.errorbar(i, Wmunu_280_500[list(Wmunu_280_500.keys())[i]]["R_MC"], 
                 yerr=Wmunu_280_500[list(Wmunu_280_500.keys())[i]]["sigma_R_MC"], fmt="oc", linestyle="-")
    
# pt 500-1000
for i in range(21):
    plt.errorbar(i, Wmunu_500_1000[list(Wmunu_500_1000.keys())[i]]["R_MC"], 
                 yerr=Wmunu_500_1000[list(Wmunu_500_1000.keys())[i]]["sigma_R_MC"], fmt="om", linestyle="-")
    
# pt 1000+
for i in range(21):
    plt.errorbar(i, Wmunu_1000[list(Wmunu_1000.keys())[i]]["R_MC"], 
                 yerr=Wmunu_1000[list(Wmunu_1000.keys())[i]]["sigma_R_MC"], fmt="oy", linestyle="-")
    
plt.axis([-1, 21, -5, 5])
plt.xticks(range(len(list(Wmunu_280_500.keys()))), list(Wmunu_280_500.keys()))
```

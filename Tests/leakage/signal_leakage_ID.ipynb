```python
import yaml
import matplotlib.pyplot as plt
import math
from os import listdir
```

```python
Wgamma_list = listdir("runID/Wgamma")
Wgamma_list.sort()
Wgamma_ID = [ int(i.split(".")[0]) for i in Wgamma_list]

Wgamma = []

for i in Wgamma_list:
    with open("runID/Wgamma/" + i, "r") as f:
        Wgamma.append(yaml.load(f, Loader=yaml.SafeLoader))
```

```python
Zgamma_list = listdir("runID/Zgamma")
del Zgamma_list[0]
Zgamma_list.sort()

Zgamma_ID = [ int(i.split(".")[0]) for i in Zgamma_list]

Zgamma = []

for i in Zgamma_list:
    with open("runID/Zgamma/" + i, "r") as f:
        Zgamma.append(yaml.load(f, Loader=yaml.SafeLoader))
```

```python
CR = list(Wgamma[0].keys())
regions = list(Wgamma[0]["SR"].keys())
regions_ = [ cr + " - " + region for cr in CR for region in regions if cr!="gammajetCR"]
regions_ = regions_ + ["gammajetCR"] 
```

# Wgamma

### c1

```python
plt.figure(num=None, figsize=(30, 12), dpi=80, facecolor='w', edgecolor='k')
colors = ['y', 'violet', 'g', 'c', 'm', 'b', 'k', 'orange', 'r' ]

Wgamma_process = ["Wenu", "Wmunu", "Wtaunu"]
Wgamma_slice = ["pt 35-70", "pt 70-140", "pt 140+"]


for i in range(len(Wgamma)):
    Na = [ Wgamma[i][CR[j]][region]["Na"] for j in range(4) for region in regions]
    Nb = [ Wgamma[i][CR[j]][region]["Nb"] for j in range(4) for region in regions]
    Ma = [ Wgamma[i][CR[j]][region]["Ma"] for j in range(4) for region in regions]
    Mb = [ Wgamma[i][CR[j]][region]["Mb"] for j in range(4) for region in regions]
    
    Na.append(Wgamma[i][CR[4]]["Na"])
    Nb.append(Wgamma[i][CR[4]]["Nb"])
    Ma.append(Wgamma[i][CR[4]]["Ma"])
    Mb.append(Wgamma[i][CR[4]]["Mb"])
    
    c1 = [ 0 for _ in range(21)]
    sigma_c1 = [ 0 for _ in range(21)]
    for j in range(21):
        if Na[j]>0 and Nb[j]>0:
            c1[j] = Nb[j]/Na[j]
            sigma_c1[j] = c1[j]*math.sqrt((1/Na[j]) + (1/Nb[j]))
    
    plt.errorbar([ j + int(i/3)*0.1 for j in range(21)], c1, yerr=sigma_c1, fmt="o", color=colors[i], linestyle="-", linewidth=1, label=Wgamma_process[int(i/3)] + " " + Wgamma_slice[i%3])
    
plt.xticks(range(len(regions_)), regions_)
plt.legend(fontsize="xx-large")
plt.title("Wgamma c1", fontsize=30)
plt.ylim(0, 0.5)
```

# Merge Wgamma

```python
for i in range(len(Wgamma)):
    if Wgamma_process[int(i/3)] != "Wenu":
        Na = [ Wgamma[i]["SR"][region]["Na"] + Wgamma[i]["1muCR"][region]["Na"] for region in regions if region != "gammajetCR"]
        Nb = [ Wgamma[i]["SR"][region]["Nb"] + Wgamma[i]["1muCR"][region]["Nb"] for region in regions if region != "gammajetCR"]
        Ma = [ Wgamma[i]["SR"][region]["Ma"] + Wgamma[i]["1muCR"][region]["Ma"] for region in regions if region != "gammajetCR"]
        Mb = [ Wgamma[i]["SR"][region]["Mb"] + Wgamma[i]["1muCR"][region]["Mb"] for region in regions if region != "gammajetCR"]

c1_W = [0 for _ in range(5)]
sigma_c1_W = [0 for _ in range(5)]
c2_W = [0 for _ in range(5)]
sigma_c2_W = [0 for _ in range(5)]
c3_W = [0 for _ in range(5)]
sigma_c3_W = [0 for _ in range(5)]

for i in range(5):
    if Na[i]>0 and Nb[i]>0 and Ma[i]>0 and Mb[i]>0:
        c1_W[i] = Nb[i]/Na[i]
        sigma_c1_W[i] = c1_W[i]*math.sqrt((1/Na[i]) + (1/Nb[i]))
        c2_W[i] = Ma[i]/Na[i]
        sigma_c2_W[i] = c2_W[i]*math.sqrt((1/Na[i]) + (1/Ma[i]))
        c3_W[i] = Mb[i]/Na[i]
        sigma_c3_W[i] = c3_W[i]*math.sqrt((1/Na[i]) + (1/Mb[i]))
```

```python
import yaml
import matplotlib.pyplot as plt
import math
```

```python
with open("noMET/dati_Wgamma.txt", "r") as f:
    Wgamma = yaml.load(f, Loader=yaml.SafeLoader)
with open("noMET/dati_Zgamma.txt", "r") as f:
    Zgamma = yaml.load(f, Loader=yaml.SafeLoader)
with open("noMET/dati_Znunugamma.txt", "r") as f:
    Znunugamma = yaml.load(f, Loader=yaml.SafeLoader)
with open("noMET/dati_gammajets.txt", "r") as f:
    gammajets = yaml.load(f, Loader=yaml.SafeLoader)
```

```python
with open("noMET/dati_Wgamma_50.txt", "r") as f:
    Wgamma_50 = yaml.load(f, Loader=yaml.SafeLoader)
with open("noMET/dati_Zgamma_50.txt", "r") as f:
    Zgamma_50 = yaml.load(f, Loader=yaml.SafeLoader)
with open("noMET/dati_Znunugamma_50.txt", "r") as f:
    Znunugamma_50 = yaml.load(f, Loader=yaml.SafeLoader)
with open("noMET/dati_gammajets_50.txt", "r") as f:
    gammajets_50 = yaml.load(f, Loader=yaml.SafeLoader)
```

```python
CR = list(Wgamma.keys())
regions = list(Wgamma["SR"].keys())
regions_ = [ cr + " - " + region for cr in CR for region in regions]
colors = ["b", "r", "orangered", "g", "violet"]
processi = ["Wgamma", "Zgamma", "Znunugamma", "gammajets"]
```

### Wgamma

```python
c1_W = []
sigma_c1_W = []
c2_W = []
sigma_c2_W = []
c3_W = []
sigma_c3_W = []

for i in range(len(CR)):
    for j in range(3):
        Na = Wgamma[CR[i]][regions[j]]["Na"]
        Nb = Wgamma[CR[i]][regions[j]]["Nb"]
        Ma = Wgamma[CR[i]][regions[j]]["Ma"]
        Mb = Wgamma[CR[i]][regions[j]]["Mb"]
        
        if Na>0 and Nb>0:
            c1_W.append(100*Nb/Na)
            sigma_c1_W.append(c1_W[j] * math.sqrt((1/Nb)+(1/Na)))
        else:
            c1_W.append(0)
            sigma_c1_W.append(0)
        
        if Na>0 and Ma>0:
            c2_W.append(100*Ma/Na)
            sigma_c2_W.append(c2_W[j] * math.sqrt((1/Ma)+(1/Na)))
        else:
            c2_W.append(0)
            sigma_c2_W.append(0)
            
        if Na>0 and Mb>0:
            c3_W.append(100*Mb/Na)
            sigma_c3_W.append(c3_W[j] * math.sqrt((1/Mb)+(1/Na)))
        else:
            c3_W.append(0)
            sigma_c3_W.append(0)
```

# Processi

```python
selectionCR = [[0, 1, 4], [0, 2, 3, 4], [0, 4], [0, 4]]

for j in range(4):
    
    fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharex=True)
    fig.set_size_inches(18.5, 5.5, forward=True)
    fig.suptitle(processi[j], fontsize=16)
    
    

    for i in selectionCR[j]:
        
        c1 = [c1_[j][k] for k in range(3*i, 3*i+3)]
        sigma_c1 = [sigma_c1_[j][k] for k in range(3*i, 3*i+3)]
        c2 = [c2_[j][k] for k in range(3*i, 3*i+3)]
        sigma_c2 = [sigma_c2_[j][k] for k in range(3*i, 3*i+3)]
        c3 = [c3_[j][k] for k in range(3*i, 3*i+3)]
        sigma_c3 = [sigma_c3_[j][k] for k in range(3*i, 3*i+3)]
        
        ax1.errorbar([l+(i/100) for l in range(3)], c1, yerr=sigma_c1, fmt="o", color=colors[i], linestyle="-", linewidth=2, label=CR[i])
        ax2.errorbar([l+(i/100) for l in range(3)], c2, yerr=sigma_c2, fmt="o", color=colors[i], linestyle="-", linewidth=2, label=CR[i])
        ax3.errorbar([l+(i/100) for l in range(3)], c3, yerr=sigma_c3, fmt="o", color=colors[i], linestyle="-", linewidth=2, label=CR[i])




    ax1.set_xticks(range(len(regions)), regions)
    ax1.legend()
    ax1.grid(linestyle="--", linewidth=1)
    ax1.set_ylabel("c1 [%]         ", rotation=0, fontsize=20)

    ax2.set_xticks(range(len(regions)), regions)
    ax2.legend()
    ax2.grid(linestyle="--", linewidth=1)
    ax2.set_ylabel("c2 [%]         ", rotation=0, fontsize=20)

    plt.xticks(range(len(regions)), regions)
    ax3.legend()
    ax3.grid(linestyle="--", linewidth=1)
    ax3.set_ylabel("c3 [%]         ", rotation=0, fontsize=20)
    #ax3.set_ylim([0, 3])

    fig.subplots_adjust(wspace=0.4)


```

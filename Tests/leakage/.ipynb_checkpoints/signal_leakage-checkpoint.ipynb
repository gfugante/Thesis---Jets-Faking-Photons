```python
import yaml
import matplotlib.pyplot as plt
import math
```

```python
with open("Wgamma_pesati.txt", "r") as f:
    Wgamma_pesati = yaml.load(f, Loader=yaml.SafeLoader)
with open("Zgamma_pesati.txt", "r") as f:
    Zgamma_pesati = yaml.load(f, Loader=yaml.SafeLoader)
with open("Znunugamma_pesati.txt", "r") as f:
    Znunugamma_pesati = yaml.load(f, Loader=yaml.SafeLoader)
with open("Wgamma_non_pesati.txt", "r") as f:
    Wgamma_non_pesati = yaml.load(f, Loader=yaml.SafeLoader)
with open("Zgamma_non_pesati.txt", "r") as f:
    Zgamma_non_pesati = yaml.load(f, Loader=yaml.SafeLoader)
with open("Znunugamma_non_pesati.txt", "r") as f:
    Znunugamma_non_pesati = yaml.load(f, Loader=yaml.SafeLoader)
```

```python
regions = list(Wgamma_pesati.keys())
x = [i for i in range(21)]
```

# Signal Leakage

## Eventi pesati

```python

Wgamma_c1           = [Wgamma_pesati[i]["c1"] for i in regions]
Wgamma_sigma_c1     = [Wgamma_pesati[i]["sigma_c1"] for i in regions]
Zgamma_c1           = [Zgamma_pesati[i]["c1"] for i in regions]
Zgamma_sigma_c1     = [Zgamma_pesati[i]["sigma_c1"] for i in regions]
Znunugamma_c1       = [Znunugamma_pesati[i]["c1"] for i in regions]
Znunugamma_sigma_c1 = [Znunugamma_pesati[i]["sigma_c1"] for i in regions]



plt.figure(num=None, figsize=(30, 12), dpi=80, facecolor='w', edgecolor='k')

plt.errorbar(x, Wgamma_c1, yerr=Wgamma_sigma_c1, fmt="or", linestyle="-", linewidth=2, label="Wgamma")
plt.errorbar(x, Zgamma_c1, yerr=Zgamma_sigma_c1, fmt="og", linestyle="-", linewidth=2, label="Zgamma")
plt.errorbar(x, Znunugamma_c1, yerr=Znunugamma_sigma_c1, fmt="ob", linestyle="-", linewidth=2, label="Znunugamma")


plt.xticks(range(len(regions)), regions)
plt.legend(fontsize="xx-large")
plt.title("Signal Leakage c1", fontsize=30)


# Wgamma
print("Wgamma")
num = 0;
den = 0;

for i in range(21):
    if Wgamma_sigma_c1[i] is not 0:
        num = num + Wgamma_c1[i]/pow(Wgamma_sigma_c1[i], 2)
        den = den + 1/pow(Wgamma_sigma_c1[i], 2)

c1_p = num/den
sigma_c1_p = math.sqrt(1/den)
print("  c1 = " + repr(round(100*c1_p, 4)) + " +- " + repr(round(100*sigma_c1_p, 4)) + " %")


# Zgamma
print("Zgamma")
num = 0;
den = 0;

for i in range(21):
    if Zgamma_sigma_c1[i] is not 0:
        num = num + Zgamma_c1[i]/pow(Zgamma_sigma_c1[i], 2)
        den = den + 1/pow(Zgamma_sigma_c1[i], 2)

c1_p = num/den
sigma_c1_p = math.sqrt(1/den)
print("  c1 = " + repr(round(100*c1_p, 4)) + " +- " + repr(round(100*sigma_c1_p, 4)) + " %")

# Znunugamma
print("Znunugamma")
num = 0;
den = 0;

for i in range(21):
    if Znunugamma_sigma_c1[i] is not 0:
        num = num + Znunugamma_c1[i]/pow(Znunugamma_sigma_c1[i], 2)
        den = den + 1/pow(Znunugamma_sigma_c1[i], 2)

c1_p = num/den
sigma_c1_p = math.sqrt(1/den)
print("  c1 = " + repr(round(100*c1_p, 4)) + " +- " + repr(round(100*sigma_c1_p, 4)) + " %")

```

## Merging

## Pesati

```python
Na = [Wgamma_pesati[i]["Na_s"] + Zgamma_pesati[i]["Na_s"] + Znunugamma_pesati[i]["Na_s"] for i in regions]
Nb = [Wgamma_pesati[i]["Nb_s"] + Zgamma_pesati[i]["Nb_s"] + Znunugamma_pesati[i]["Nb_s"] for i in regions]
Ma = [Wgamma_pesati[i]["Ma_s"] + Zgamma_pesati[i]["Ma_s"] + Znunugamma_pesati[i]["Ma_s"] for i in regions]
Mb = [Wgamma_pesati[i]["Mb_s"] + Zgamma_pesati[i]["Mb_s"] + Znunugamma_pesati[i]["Mb_s"] for i in regions]


    
print(Na)
c1 = [0 for _ in range(21)]
sigma_c1 = [0 for _ in range(21)]
c2 = [0 for _ in range(21)]
sigma_c2 = [0 for _ in range(21)]
c3 = [0 for _ in range(21)]
sigma_c3 = [0 for _ in range(21)]




c1 = [ Nb[i]/Na[i] for i in range(21) if Na[i]>0 and Nb[i]>0 ]
sigma_c1 = [ c1[i] * math.sqrt((1/Na[i]) + (1/Nb[i])) for i in range(21) if Na[i]>0 and Nb[i]>0 ]

c2 = [ Ma[i]/Na[i] for i in range(21) if Na[i]>0 and Ma[i]>0 ]
sigma_c2 = [ c2[i] * math.sqrt((1/Na[i]) + (1/Ma[i])) for i in range(21) if Na[i]>0 and Ma[i]>0 ]

c3 = [ Mb[i]/Na[i] for i in range(21) if Na[i]>0 and Mb[i]>0 ]
sigma_c3 = [ c3[i] * math.sqrt((1/Na[i]) + (1/Mb[i])) for i in range(21) if Na[i]>0 and Mb[i]>0 ]

```

```python
plt.figure(num=None, figsize=(30, 12), dpi=80, facecolor='w', edgecolor='k')

plt.errorbar(x, c1, yerr=sigma_c1, fmt="or", linestyle="-", linewidth=2, label="c1")


plt.xticks(range(len(regions)), regions)
plt.legend(fontsize="xx-large")
plt.title("Signal Leakage c1", fontsize=30)

# media pesata sugli errori

num = 0;
den = 0;

for i in range(21):
    if sigma_c1[i] is not 0:
        num = num + c1[i]/pow(sigma_c1[i], 2)
        den = den + 1/pow(sigma_c1[i], 2)

c1_p = num/den
sigma_c1_p = math.sqrt(1/den)
print("c1 = " + repr(round(100*c1_p, 4)) + " +- " + repr(round(100*sigma_c1_p, 4)) + " %")


plt.hlines(c1_p+sigma_c1_p, 0, 21, colors='gray', linestyles='solid', label='', linewidth=0.5)
plt.hlines(c1_p-sigma_c1_p, 0, 21, colors='gray', linestyles='solid', label='', linewidth=0.5)

```

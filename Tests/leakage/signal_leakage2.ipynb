```python
import yaml
import matplotlib.pyplot as plt
import math
```

```python
with open("CaloTrack/dati_Wgamma.txt", "r") as f:
    Wgamma = yaml.load(f, Loader=yaml.SafeLoader)
with open("CaloTrack/dati_Zgamma.txt", "r") as f:
    Zgamma = yaml.load(f, Loader=yaml.SafeLoader)
with open("CaloTrack/dati_Znunugamma.txt", "r") as f:
    Znunugamma = yaml.load(f, Loader=yaml.SafeLoader)
with open("CaloTrack/dati_gammajets.txt", "r") as f:
    gammajets = yaml.load(f, Loader=yaml.SafeLoader)
```

```python
CR = Wgamma.keys()
regions = list(Wgamma["SR"].keys())
regions_ = [ cr + " - " + region for cr in CR for region in regions if cr!="gammajetCR"]
regions_ = regions_ + ["gammajetCR"] 
track_limit = 0.1
print (regions_)

def w_avarage(x, sigma):
    num = 0
    den = 0
    for i in range(len(x)):
        num = num + x[i]/pow(sigma[i],2)
        den = den + 1/pow(sigma[i],2)
    return [num/den, math.sqrt(1/den)]
        
```

```python
# i vari track_limit non influenzano molto il leakage, scelgo quello che elimina quella striscia strana di eventi
# calo isolati ma non track isolati -> track_limit = track_limit

# mergio la SR e la 1muCR
Na_W = [0 for _ in range(5)]
Nb_W = [0 for _ in range(5)]
Ma_W = [0 for _ in range(5)]
Mb_W = [0 for _ in range(5)]



for i in range(5):
    Na_W[i] = Wgamma["SR"][regions[i]][track_limit]["Na"] + Wgamma["1muCR"][regions[i]][track_limit]["Na"]
    Nb_W[i] = Wgamma["SR"][regions[i]][track_limit]["Nb"] + Wgamma["1muCR"][regions[i]][track_limit]["Nb"]
    Ma_W[i] = Wgamma["SR"][regions[i]][track_limit]["Ma"] + Wgamma["1muCR"][regions[i]][track_limit]["Ma"]
    Mb_W[i] = Wgamma["SR"][regions[i]][track_limit]["Mb"] + Wgamma["1muCR"][regions[i]][track_limit]["Mb"]
```
